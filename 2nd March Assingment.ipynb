# 1.

Matplotlib is a popular Python library for creating static, animated, and interactive visualizations in a variety of formats. It's widely used for data visualization and plotting in scientific computing, machine learning, and other domains.

Line Plot:
It's used to visualize the relationship between two continuous variables by connecting data points with straight lines.

Scatter Plot:
This plot is used to display individual data points on a two-dimensional plane. It's great for visualizing the distribution and relationships between two variables.

Bar Plot:
Bar plots are useful for comparing the values of different categories or groups. They represent data using rectangular bars with lengths proportional to the values they represent.

Histogram:
Histograms are used to visualize the distribution of a single variable. It displays the frequencies of different intervals, making it easy to identify patterns like skewness or central tendency.

Pie Chart:
Pie charts are useful for showing the proportions of different categories in a dataset. Each slice of the pie represents a different category, and the size of each slice corresponds to the proportion of the whole.

# 2.

A scatter plot is a type of plot that displays individual data points on a two-dimensional plane, with one variable on the x-axis and another on the y-axis. It's great for visualizing the relationship or distribution between two variables.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y, color='blue', alpha=0.7)  # alpha adjusts transparency

# Add title, xlabel, and ylabel
plt.title('Scatter Plot Example')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Show the plot
plt.show()
```

# 3.

The subplot() function in Matplotlib is used to create multiple plots within the same figure. It allows you to organize and display multiple plots in a grid, making it easier to compare and analyze different datasets

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for four lines
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st plot
plt.plot(x, y1, label='Line 1')
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd plot
plt.plot(x, y2, label='Line 2')
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)  # 2 rows, 2 columns, 3rd plot
plt.plot(x, y3, label='Line 3')
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)  # 2 rows, 2 columns, 4th plot
plt.plot(x, y4, label='Line 4')
plt.title('Line Plot 4')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```

# 4.

A bar plot is a visualization that represents categorical data with rectangular bars. The length of each bar corresponds to the value of the data it represents. Bar plots are commonly used to compare the values of different categories or groups.

A horizontal bar plot is similar to a bar plot, but the bars are displayed horizontally, making it easier to compare the lengths of bars for different categories.

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.figure(figsize=(10, 6))
plt.bar(company, profit, color='skyblue')
plt.title('Profit by Company (Bar Plot)')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(10, 6))
plt.barh(company, profit, color='violet')
plt.title('Profit by Company (Horizontal Bar Plot)')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.show()
```

# 5.

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum of a dataset.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot of box1 and box2')
plt.xlabel('Box')
plt.ylabel('Values')
plt.show()
```
