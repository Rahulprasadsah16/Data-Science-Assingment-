```python
# First we need to convert both 'price' and 'type' in numerice type.
```

```python
df1['Price'].unique()
```

```python
chars_to_remove = ['$',',','+']
columns_to_clean = ['Installs','Price']
for item in chars_to_remove:
    for col in columns_to_clean:
        df1[col] = df1[col].str.replace(item,'')
```

```python
df1['Price'].unique()
```

```python
df1 = df1[df1['Price'] != 'Everyone'] 
```

```python
df1['Type'].unique()
```

```python
df1 = df1[df1['Type'] != 'nan'] 
```

```python
df1['Type'].describe()
```

```python
df1['Type'] = df1['Type'].map({'Free':0, 'Paid':1,np.nan:1})
```

```python
# Create a bar chart

plt.figure(figsize=(10, 6))
df1.plot(kind='bar', color='skyblue')
plt.title('Average Prices by App Type')
plt.xlabel('Type')
plt.ylabel('Price')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
# Remove the string "Free" at position 10472 in the Installs column

df1.loc[10472, 'Installs'] = df1.loc[10472, 'Installs'].replace('Free', '')
```

```python
# Create a frequency table of app installs
install_freq = df1['Installs'].value_counts()

# Convert the 'Installs' column to numeric after removing '+' and ',' characters
df1['Installs'] = df1['Installs'].str.replace('+', '')
df1['Installs'] = df1['Installs'].str.replace(',', '')
df1['Installs'] = pd.to_numeric(df1['Installs'])

# Sort the frequency table in descending order
install_freq_sorted = install_freq.sort_values(ascending=False)

# Display the top 10 most popular apps
top_10_apps = install_freq_sorted.head(10)
print("Top 10 Most Popular Apps:")
print(top_10_apps) 
```
